```python
dict = {}
readFile = open('out_text.txt', 'r')
for line in readFile.readlines():
    words = (line.lower()).split()
    for word in words:
        if word.isalpha():
            if word not in dict.keys():
                dict[word] = 1
            if word in dict.keys():
                dict[word] += 1
print(dict)
readFile.close()
```

```python
def getMax(dict):
    max = 0
    maxW = ''
    for word in dict:
        if max < dict[word]:
            max = dict[word]
            maxW = word
    return maxW
sortedDict = sorted(dict.items(), key = lambda kv:(kv[1], kv[0]))
print(sortedDict)
```

```python
tempDict = dict.copy()
frequentWords = []
for i in range(500):
    word = getMax(tempDict)
    if word.isalpha():
        frequentWords.append(word)
        i = i + 1
    del tempDict[word]
```

```python
print(frequentWords)
print(dict['rain'])
```

```python
outfile = open('words.txt', 'w')
for word in frequentWords:
    outfile.write(word)
    outfile.write('\n')
outfile.close()
```
