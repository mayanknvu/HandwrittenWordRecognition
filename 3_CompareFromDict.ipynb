```python
import PredictWord
from fuzzywuzzy import fuzz 
from fuzzywuzzy import process
```

```python
def FindClosest(words, pred):
    highest = 0
    allClose = []
    for word in words:
        ratio = fuzz.ratio(word, pred)
        if highest < ratio:
            highest = ratio
    for word in words:
        if fuzz.ratio(word, pred) == highest:
            allClose.append(word)
    return allClose
```

```python
def getList():
    words = []
    with open('ocr/words-linux.txt') as infile:
        for line in infile.readlines():
            temp_words = line.split()
            words.append(temp_words[0])
    return words
```

```python
def RunProgram(image):
    prediction = PredictWord.PredictString(image)
    print('PREDICTION WITHOUT CASE: ' + prediction.lower())
    words = getList()
    closeMatch = FindClosest(words, prediction.lower())
    print('all close matches are: ')
    for word in closeMatch:
        print(word)
```

```python
RunProgram('test.png')
```
